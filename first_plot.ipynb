## First Plot: learning how to draw plots in numpy

Здесь я хочу разобраться  в том, как работают, квадратные уравнения, а также уравнения прямых.

$$ax^2+bx+c=0 $$

Важно знать о **дискриминанте**, его формула:

$$D = b^2 - 4ac $$

```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sym
from valeras_library.valeras_functions import discriminant
```

```python
x = np.linspace(-50, 50, 1000)

a = .25
b = 0
c = -4

d = discriminant(a=1, b=20, c=3)
print(f"Discriminant for a={a}, b={b}, c={c}: {d}")

y = a*x**2 + b*x + c


plt.plot(x, y, 'c.', markersize=10, alpha=.3)
plt.plot([-25, 20], [0, 0], 'k--')
plt.plot([0, 0], [-5, 20], 'k--')

plt.axis('square')
plt.axis([-40, 40, -10, 40])
plt.gcf().set_size_inches(10, 10)
plt.grid()
plt.show()
```

```python
x_sym = sym.symbols('x')
expr = x_sym**2+2*x_sym-1

solutions = sym.solve(expr)

x = np.linspace(-10, 10, 100)
y = x**2 + 2*x -1

plt.plot(x, y, 'g-', markersize=2)
plt.plot([-25, 20], [0, 0], 'k--', alpha=.5)
plt.plot([0, 0], [-5, 20], 'k--', alpha=.5)

for solution in solutions:
    sol = np.float64(solution)
    print(sol)
    plt.plot([sol, sol], [-5, 20], 'r--', linewidth=1)

plt.axis('square')
plt.axis([-5, 5, -5, 5])
plt.gcf().set_size_inches(7, 7)
plt.grid()
plt.show()
```

```python
x = np.linspace(-10, 10, 100)
y1 = x**2-1
y2 = x**2

plt.plot(x, y1, 'c-', markersize=2)
plt.plot(x, y2, 'g-', markersize=2)
plt.plot([-25, 20], [0, 0], 'k--')
plt.plot([0, 0], [-5, 20], 'k--')

# plt.plot([-1+np.sqrt(2), -1+np.sqrt(2)], [-5, 20], 'r--')
# plt.plot([-np.sqrt(2)-1, -1-np.sqrt(2)], [-5, 20], 'r--')

plt.axis('square')
plt.axis([-5, 5, -5, 5])
plt.gcf().set_size_inches(7, 7)
plt.grid()
plt.show()
```
